When switching between frequencies with the DDS (334MHz, 337MHz, 340MH) we see spectral density between 50kHz and 1MHz on the spectrum analyzer. Each frequency is on for between (1-10)$\mu s$ I want to try and reproduce the signal and then take the FFT

```python
%matplotlib notebook
T = 1e-10 # Time step
N = 1e5 # Number of points, 

f =  [334.0e6, 337.0e6, 340.0e6, 0]
t = np.array([])
y = np.array([])
cycles = 6
for i in range(cycles*len(f)):
    t1 = np.linspace(i*N*T, N*T*(i+1), N)
    t = np.append(t, t1)
    y = np.append(y, np.sin(2*np.pi*f[i%4]*t1))

#print t[2e4:2.1e4]
#plot(t[10.99e5:11.01e5],y[10.99e5:11.01e5])

#show()
fft = np.fft.fft(y)
freq = np.linspace(0,1./2./T,N*cycles*len(f)/2)
fft = abs(fft)
#print freq[-2]
plot(freq,fft[0:N*cycles*len(f)/2])


T = 1e-10 # Time step
N = .5e5 # Number of points, 

f =  [334.0e6, 337.0e6, 340.0e6, 0]
t = np.array([])
y = np.array([])
cycles = 6
for i in range(cycles*len(f)):
    t1 = np.linspace(i*N*T, N*T*(i+1), N)
    t = np.append(t, t1)
    y = np.append(y, np.sin(2*np.pi*f[i%4]*t1))

#print t[2e4:2.1e4]
#plot(t[10.99e5:11.01e5],y[10.99e5:11.01e5])

#show()
fft = np.fft.fft(y)
freq = np.linspace(0,1./2./T,N*cycles*len(f)/2)
fft = abs(fft)
#print freq[-2]
plot(freq,fft[0:N*cycles*len(f)/2])



xlim([3e4,1.7e5])
ylim([0,100])
show()
```
