Scan of 455nm line for ba-138

```python
# Read in the data
rcParams['axes.linewidth'] = 4
from scipy.optimize import curve_fit

center_freq = 658.116320

data = np.loadtxt('H:/Data/2017.dir/11.dir/2017_11_07.dir/00002 - Line_Scan.csv', delimiter = ',')
data1 = np.loadtxt('H:/Data/2017.dir/11.dir/2017_11_07.dir/00003 - Line_Scan.csv', delimiter = ',')


frequency = np.concatenate((data[:,0],data1[:,0]),axis = 0)
counts = np.concatenate((data[:,1], data1[:,1]), axis = 0)
counts = counts/max(counts)


# define fit function
def gauss(x, A, mu, sigma, C):
    return A*np.exp(-(x-mu)**2/(2.*sigma**2)) + C

# Fit
fit_g, var_g = curve_fit(gauss,frequency, counts)

print fit_g[1]
print "Center freq: ", "{:6f}".format(center_freq + fit_g[0]*1e-6 + 40e-6)

plot_freq = np.linspace(frequency[0],frequency[-1],1000)
fit_data_g = gauss(frequency, fit_g[0],fit_g[1],fit_g[2],fit_g[3])

plot(frequency,fit_data_g,color = 'r', linewidth = 4)

plot(frequency,counts,'o', markersize = 16)
xlabel('Frequency (GHz)', fontsize = 46)
ylabel('counts', fontsize = 46)
xlim([-400,600])
tick_params(labelsize = 44)
tick_params(width=5, length=10)
show()
```
