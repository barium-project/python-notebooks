## Shelving and 455nm Spectroscopy of Ba-133

First experiments were taking lines scans of the 455 transitions in ba-133. We scanned both lines $|S_{1/2},F =1> \leftrightarrow |P_{3/2}, F = 2>$ and the $|S_{1/2},F =1> \leftrightarrow |P_{3/2}, F = 1>$

# $|S_{1/2},F =1> \leftrightarrow |P_{3/2}, F = 2>$ 

```python
# Read in the data
rcParams['axes.linewidth'] = 4
from scipy.optimize import curve_fit
rcParams["figure.figsize"] = 15, 10

data = np.loadtxt('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/00025 - shelving133_prob.csv', delimiter = ',')

f0 = 658.118744 # guess
frequency = (data[:,0] - f0)*1e6 # in MHz
prob = data[:,1]


# define fit function
def gauss(x, A, mu, sigma, C):
    return A*np.exp(-(x-mu)**2/(2.*sigma**2)) + C

def lorentz(x, A, mu, sigma, C):
    return A*sigma/np.pi/((x-mu)**2+sigma**2) + C 

# Intial guess
p0 = [.1 ,0 ,100, 0]
p1 = [1 ,0 ,50, 0]



# Fit
fit_g, var_g = curve_fit(gauss,frequency, prob, p0)
print fit_g
print var_g[1][1], np.sqrt(var_g[1][1])

fit_l, var_l = curve_fit(lorentz,frequency, prob, p1)

print fit_l
print var_l[1][1], np.sqrt(var_l[1][1])



fit_data_g = gauss(frequency, fit_g[0],fit_g[1],fit_g[2],fit_g[3])
fit_data_l = lorentz(frequency, fit_l[0],fit_l[1],fit_l[2],fit_l[3])

center = f0 + fit_l[1]*1e-6

print 'Center Frequency ', '{:.7f}'.format(center), 'THz'


plot(frequency,prob, 'o', markersize = 18)
plot(frequency,fit_data_g,color = 'g')
plot(frequency,fit_data_l, color = 'r', linewidth = 4.5)

#title('Barium-133 P-1/2 Hyperfine Splitting', fontsize = 24)
xlabel('Frequency (MHz)', fontsize = 46)
ylabel('Probability', fontsize = 46)
tick_params(labelsize = 44)
tick_params(width=5, length=10)
show()

#savefig('Z:/Group_Share/Barium/Data/2017.dir/02.dir/2017_02_15.dir/ba_133_p_1_2.png', bbox_inches='tight')
#savefig('Z:/Group_Share/Barium/Data/2017.dir/02.dir/2017_02_15.dir/ba_133_p_1_2.pdf', bbox_inches='tight')
#savefig('Z:/Group_Share/Barium/Data/2017.dir/02.dir/2017_02_15.dir/ba_133_p_1_2.eps', bbox_inches='tight', format='eps', dpi = 1000)
```

## Shelving 133

```python
# Read in the data
rcParams['axes.linewidth'] = 4
from scipy.optimize import curve_fit
rcParams["figure.figsize"] = 15, 10
data = np.loadtxt('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/00260 - shelving133_prob.csv', delimiter = ',')
data1 = np.loadtxt('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/00263 - shelving133_prob.csv', delimiter = ',')

t = np.concatenate((data[:,0], data1[:,0]), axis=0)

prob =  np.concatenate((data[:,1], data1[:,1]), axis=0)


# define fit function
def shelve(t, A, tau, C):
    return A*(1-np.exp(-t/tau)) + C


# Intial guess
p0 = [.87 ,5 , 0]


# Fit
fit, var = curve_fit(shelve,t, prob, p0)
print fit
print var[0][0], var[1][1]



fit_data = shelve(t, fit[0],fit[1],fit[2])



plot(t,prob, 'o', markersize = 18)
plot(t,fit_data,color = 'r', linewidth = 4.5)

#title('Barium-133 P-1/2 Hyperfine Splitting', fontsize = 24)
xlabel('Time (us)', fontsize = 46)
ylabel('Probability', fontsize = 46)
tick_params(labelsize = 44)
tick_params(width=5, length=10)
ylim([0,1])
show()

#savefig('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/ba_133_shelving.png', bbox_inches='tight')
#savefig('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/ba_133_shelving.pdf', bbox_inches='tight')
#savefig('Z:/Group_Share/Barium/Data/2018.dir/01.dir/2018_01_10.dir/ba_133_shelving.eps', bbox_inches='tight', format='eps', dpi = 1000)
```
