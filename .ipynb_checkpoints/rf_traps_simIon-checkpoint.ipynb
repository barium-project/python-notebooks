## Loading Electrodes into Simion

1. Make a CAD drawing of the object you're simulating. Make sure to reference the CAD origin when placing electrodes.
2. Each electrode in the CAD drawing needs to be saved as a .STL file. The easiest way to do this is place all parts in an assembly, then suppress all but one. Save the files with the following naming convention "electrode-%.STL" where the % is replaced by the electrode number (i.e. 4 for the fourth electrode in your drawing. Also make sure you know what units your SAVING in. Inventor will tell you your drawing is in mm, but save it in cm. Click options in Inventor,go to "Save as", "Save Copy as". Click "options" at the bottom to choose units. Save all files in the same spot.
3. Open SimIon and click OK.
4. Click SL tools and then STL->PA on the left of the window that pops up.
5. Click browse on the input file and select your first electrode. Then replace the electrode number with a "%". i.e "electrode-1.stl" goes to "electrode-%.stl". This will load in all files. 
6. In the region scale add the scale factor.
7. In grid point count for "x" change the number to change the number of points. y and z will automatically change.
8. Check the view output box and then click "convert". If the electrodes look good exit the window.
9. In the main window in the Potential Arrays section highlight the new file and select "refine".
10. After refine is finished select "View/Load" workbench

## Programs for Time Dependent Voltages

Use a program to add time dependence and intialize voltages. Programming is done in .lua language. Electrodes are numbered by the number assigned in the .stl file. In the workbench click the particles tab and select the "User Program" button. When asked to create a file click yes. An example file is given:

simion.workbench_program()

-- all units in volts

adjustable focus= 0.0

adjustable RFfreq_Hz = 2.7648e6

adjustable DC_bias_pos = 2.8

adjustable DC_bias_neg = -1.9

adjustable RFamp = 32.0


local Omega = 2 x math.pi x RFfreq_Hz x 1e-6 -- Simion keeps time in microseconds

function segment.init_p_values()

	adj_elect01 = DC_bias_pos
	adj_elect03 = DC_bias_pos

	adj_elect02 = DC_bias_neg
	adj_elect04 = DC_bias_neg
	adj_elect05 = focus
end

function segment.fast_adjust()

	local v = RFamp*sin(Omega*ion_time_of_flight)

	adj_elect01 = DC_bias_pos + v
	adj_elect03 = DC_bias_pos + v 

	adj_elect02 = DC_bias_neg - v
	adj_elect04 = DC_bias_neg - v

end

### Mathieu Parameters

Stability is determined by the "a" and "q" parameters of the Mathieu equations. This is for a 2-d quadrupole filter. Their definitions are as follows:

$a = \frac{8eV_{DC}}{mr_{0}^2\omega^2}$ <br><br>
$q = \frac{4eV_{RF}}{mr_{0}^2\omega^2}$ <br><br>
where $r_0$ is the distance from the center to the hyperbolic rod and $\omega$ is the RF drive frequency. These values of a and q are defined such that $\phi$ is the voltage on the electrodes. Other definitions have a,q, and $\phi$ divided by 2. 

#### Operating line

Both a and q are funcions of m, substituting to get 'a' as a function of 'q' gives you an equation for a straight line independent of m. This equation is :

$a = \frac{2V_{DC}}{V_{RF}}q$

All masses fall on this line with increasing mass moving towards the origin. As long as the ratio of

$\frac{2V_{DC}}{V_{RF}}$ 

stays the same, the operating line does not move. The goal for mass filtering is to arrange the line such that it just passes through the upper apex of the stability diagram. Then only a single mass will be stable. Then the mass which is stable can be adjusted by properly scaling VRF and VDC with the mass. So, for example find VDC and VRF for mass=1 such that 'a' and 'q' are in the apex of the stability diagram. Then scale as follows:

$a = \frac{n8eV_{DC}}{nmr_{0}^2\omega^2}$

$q = \frac{n4eV_{DC}}{nmr_{0}^2\omega^2}$

where $n\cdot m$ is the total mass. In this way you shift the mass you want into the upper corner (keep the same value of a and q) without changing the operating line $\frac{2V_{DC}}{V_{RF}}$.

The trap depth can be calculated for $a\ll q^2 \ll 1$. In this case the psuedopotential can be approximated as a harmonic oscillator with:

$E_0 = \frac{1}{2}m\omega^2r_{0}^2$

$\omega = \beta\frac{\Omega_{RF}}{2}$ 

$\beta = \sqrt{a+\frac{q^2}{2}}$

To understand what's going on I'm starting with a perfect hyperbola trap with an $r_0$ of 6.35mm. Christain has numerically calculated the a and q parameters and to a good approximation $a=\pm(.48q^2)$ and $q = \pm(1.09-1.2q)$. These are the theorectical values plotted below.

```python
### Calculate a and q 
e = 1.60217662e-19
r = .00635 # this is r of perfect hyperbola hyperbola
#r = .00254 # r of RGA
#omega = 2.7648e6*2*np.pi
omega = 1.1e6*2*np.pi
amu = 1.66054e-27

n = 133

VDC = .591*n # These are the voltages to make only 1amu stable in the top corner
VRF = 3.52*n # of the stability diagram for hyperbola

#VDC = .597*n # These are for the RGA
#VRF = 3.56*n


#VDC = 518
#VRF = 3224

m = n*amu

a = (8*e*VDC)/(m*r**2*omega**2)

q = (4*e*VRF)/(m*r**2*omega**2)

## Theoretical values of a and q
q_th1 = np.linspace(0,.707,100)
q_th2 = np.linspace(.707,.9,30)
a_th = .48*q_th1**2
a_th2 = 1.09-1.2*q_th2

## Find secular frequency and trap depth
beta = np.sqrt(a+q**2/2)
secular = beta*omega/2
energy = .5*m*secular**2*r**2/e


print 'a  ',a
print 'q  ', q 
print 'Operating line  ', 2*VDC/VRF
print 'DC','RF  ',VDC,VRF
print 'Secular frequency (Hz)  ',secular/2/np.pi
print 'Trap depth eV  ', energy



plot(q_th1,a_th)
plot(q_th2,a_th2)
plot(q_th1,-a_th)
plot(q_th2,-a_th2)
plot(q,a,'o',color='r')
plot(q_th1,q_th1*2*VDC/VRF,color = 'k')
plot(q_th2,q_th2*2*VDC/VRF,color = 'k')
xlabel('q',fontsize=20)
ylabel('a',fontsize=20)
tick_params(axis='both', which='major', labelsize=16)
show()
```
