Mode matching into the 493 fiber EOM has become poor. Looks like the mode coming out of the laser has changed. Will measure the beam waiste at a few locations and fit for a beam profile. 

```python
from scipy.optimize import curve_fit
# Define functions for fitting

def erf(z):
    t = 1.0 / (1.0 + 0.5 * abs(z))
    # use Horner's method
    ans = 1 - t * np.exp( -z*z -  1.26551223 +
                            t * ( 1.00002368 +
                            t * ( 0.37409196 + 
                            t * ( 0.09678418 + 
                            t * (-0.18628806 + 
                            t * ( 0.27886807 + 
                            t * (-1.13520398 + 
                            t * ( 1.48851587 + 
                            t * (-0.82215223 + 
                            t * ( 0.17087277))))))))))
    for i in range(len(z)):
        if z[i] < 0.0:
            ans[i] = -1*ans[i]
    return ans

# Define matrix for thin lens, mirror, and straight section

def thin_lens(f1):
    x = np.array([[1.,0.0],[-1./f1,1.0]])

    return x

def mirror(R):
    x = np.array([[1,0],[-2./R,1]])
    return x

def strt(d):
    x = np.array([[1,d],[0,1]])

    return x

# power function for fitting waiste
def power_func(x,w):
    return Pin/2*(1+erf(np.sqrt(2)*x/w))

def waist_func(z,d,w0):
    # The actual function assumes minimum waist at z=0. Since
    # we choose and arb coordinate system, add parameter "d"
    # to account for this. "d" is where waist is minimum.
    z0 = np.pi*n*w0**2/lam
    return np.sqrt((1+((z+d)/z0)**2)*w0**2)

def radius(z, d, w0):
    # See above comment about "d"
    z0 = np.pi*n*w0**2/lam
    r = (z+d)*(1+(z0/(z+d))**2)
    return r

# Function to find q value give R and Z
def q_value_from_beam(w2,R):
    real = 1./R
    complex = -lam/n/np.pi/w2

    return real+complex*1j

def q_value_from_q(q,Tmatrix):

    return (Tmatrix[1][0]+Tmatrix[1][1]*q)/(Tmatrix[0][0]+Tmatrix[0][1]*q)

def waist_from_q(q):

    b = -np.imag(q)
    return np.sqrt(lam/(n*np.pi*b))

def radius_from_q(q):
    a = np.real(q)
    return 1/a

```

Minimum spot size:<br>
$\Large{ w_0 = \frac{\lambda f}{\pi d}}$

Minimu spot location: <br>

$\Large{R(z) = z(1+(\frac{z_0}{z})^2)}$

Use the beam waist function to find q value and try and optimize a telescope to match the mode at the collimator.

```python
# 493 beam propagation
# mm units

coll_loc = .620
target_waist = 332.e-6
epsilon =  -0.003


# Need to create a loop to optimize beam at collimator lens
# Start by iterating the lens location.
# Define an array over range of lens locations
lens_pos = np.linspace(0.01, .400,1e4)
#lens_pos = np.array([.1])
d = np.array([-0.02]) #np.linspace(-.15,.15,1000)
# Track the difference in waiste size
waist_diff_x = np.zeros(len(lens_pos))
waist_diff_y = np.zeros(len(lens_pos))
# Track the waist at collimator
wx = np.zeros(len(lens_pos))
wy = np.zeros(len(lens_pos))
# Track radius at collimator
radius_x = np.zeros(len(lens_pos))
radius_y = np.zeros(len(lens_pos))

for i in range(len(lens_pos)):
    # First calculate the beams waist and curvature at z location
    # Values taken from fit in beam_waist.py
    x_waist =  waist_func(lens_pos[i],wofx[0], wofx[1]) #mm
    y_waist =  waist_func(lens_pos[i],wofy[0], wofy[1])
    x_radius = radius(lens_pos[i], wofx[0], wofx[1])
    y_radius = radius(lens_pos[i], wofy[0], wofy[1])

    xq = q_value_from_beam(x_waist**2, x_radius)
    yq = q_value_from_beam(y_waist**2, y_radius)

    f1 = .1
    f2 = -.030
    d1 = f1 + f2 + d[0]
    d2 = coll_loc  - d1 - lens_pos[i]

    # Next define Tmatrix
    Tlens1 = thin_lens(f1)
    Tlens2 = thin_lens(f2)
    Tprop1 = strt(d1)
    Tprop2 = strt(d2)
    #Tmatrix = np.dot(Tprop2,Tlens1)
    Tmatrix = np.dot(Tprop2,np.dot(Tlens2, np.dot(Tprop1,Tlens1)))

    # Find new q value at collimator
    xq1 = q_value_from_q(xq,Tmatrix)
    yq1 = q_value_from_q(yq,Tmatrix)

    # Find waist
    wx1 = waist_from_q(xq1)
    wy1 = waist_from_q(yq1)
    # Find radius
    rx1 = radius_from_q(xq1)
    ry1 = radius_from_q(yq1)

    wx[i] = wx1
    wy[i] = wy1

    waist_diff_x[i] = target_waist - wx1
    waist_diff_y[i] = target_waist - wy1

    radius_x[i] = rx1
    radius_y[i] = ry1

x_min = np.argmin(np.absolute(waist_diff_x))
y_min = np.argmin(np.absolute(waist_diff_y))

# Use these to calculate beam waist after asphere and then calc efficiency
wx0 = wx[x_min]
wy0 = wy[x_min]
rx0 = radius_x[x_min]
ry0 = radius_y[x_min]

print  'smallest diff in x: ', np.amin(np.absolute(waist_diff_x))
print  'smallest diff in y: ', np.amin(np.absolute(waist_diff_y))
print 'x waist at collimator: ',  wx[x_min]
print 'y waist at collimator: ', wy[x_min]
print 'x_lens_pos: ', lens_pos[x_min]
print 'y_lens_pos: ', lens_pos[y_min]
print 'x radius at collimator: ', radius_x[x_min]
print 'y radius at collimator: ', radius_y[x_min]

subplot(2,1,1)
plot(lens_pos,waist_diff_x,'b')
plot(lens_pos,waist_diff_y,'r')
#xlim([.15,.21])
subplot(2,1,2)
plot(lens_pos,radius_x,'b')
#plot(d,radius_y,'r')
show()
```

### Theoretical efficiency

For a laser with gaussian mode:<br>
$\Large{\Psi_l = e^{-(x/w_x)^2}e^{-(y/w_{mfd})^2}}$

and fiber mode:<br>

$\Large{\Psi_f = e^{-(x^2+y^2)/w_{mfd}^2}}$

the theorectical coupling efficiency is:

$\Large{4\frac{w_x w_y w_{mfd}^2}{(w_y^2 + w_{mfd}^2)(w_x^2 + w_{mfd}^2)}}$

```python
#wx0 = 329.e-6
#wy0 = 329.e-6
#rx0 = 2.2
#ry0 = 2.0

NA = .075
MFD = 3.75e-6 # want the spot size to be half this
wm = MFD/2.

f = 4.5e-3 #m focal lenght of the collimator we have
lam = 493.41e-9

print wx0,wy0, rx0, ry0
qx1 = q_value_from_beam(wx0**2, rx0) 
qy1 = q_value_from_beam(wy0**2, ry0)

# Next define Tmatrix
Tlens1 = thin_lens(f)
Tprop1 = strt(f)
Tmatrix = np.dot(Tprop1,Tlens1) 

qx2 = q_value_from_q(qx1, Tmatrix)
qy2 = q_value_from_q(qy1, Tmatrix)


wx1 = waist_from_q(qx2)
wy1 = waist_from_q(qy2)
rx2 = radius_from_q(qx2)
ry2 = radius_from_q(qy2)

print wx1, wy1, rx2, ry2

#Find coupling efficiency


eta1 = 4*(wm**2*wx1*wy1)
eta2 = (wx1**2+wm**2)
eta3 = (wy1**2+wm**2)

print wx0/f, wy0/f

eta = eta1/eta2/eta3
print eta
```

Includeing axial and radial offsets the efficiency is given by:<br>

$\eta = \eta_x\eta_y$
