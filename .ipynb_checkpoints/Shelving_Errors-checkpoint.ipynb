### Shelving Errors

It appears that the AOM is not totally off when we have it set to off during an experiment. We ran two experiments which are identical except for one the 455nm AOM is unplugged.

```python
# Need the error function for calculating probability. Grabbed this off the internet a long time ago.
# Standard erf cannot intake array of data, found this online and modified

def erf(z):
    t = 1.0 / (1.0 + 0.5 * abs(z))
    # use Horner's method
    ans = 1 - t * np.exp( -z*z -  1.26551223 +
                            t * ( 1.00002368 +
                            t * ( 0.37409196 + 
                            t * ( 0.09678418 + 
                            t * (-0.18628806 + 
                            t * ( 0.27886807 + 
                            t * (-1.13520398 + 
                            t * ( 1.48851587 + 
                            t * (-0.82215223 + 
                            t * ( 0.17087277))))))))))
    for i in range(len(z)):
        if z[i] < 0.0:
            ans[i] = -1*ans[i]
    return ans

# For fitting histograms
def gauss(x, A, mu, sigma):
    return A*np.exp(-(x-mu)**2/2/sigma**2)
```

```python
# Give the above data, what fraction of the gaussian is below our threshold?
thresh = 13
mu = fit_g[1]
sigma = fit_g[2]
print mu, sigma
x_err = (mu-thresh)/np.sqrt(2)/sigma
x_err = np.array([x_err])
len(x_err)
p = .5*(1-erf(np.array(x_err)))
print p

# what about the low end of counts

x_err = (40.-thresh)/np.sqrt(2)/np.sqrt(40.)
x_err = np.array([x_err])
len(x_err)
p = .5*(1-erf(np.array(x_err)))
print p
```
